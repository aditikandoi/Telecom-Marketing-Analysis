```python
ballist=list()
for i in range(len(data['balance'])):
    if data['balance'][i] < 20000 :
        ballist.append('0-20000')
    elif data['balance'][i] < 40000:
         ballist.append('20001-40000')
    elif data['balance'][i] < 60000:
         ballist.append('40001-60000')
    elif data['balance'][i] < 80000:
         ballist.append('60001-80000')
    elif data['balance'][i] < 100000:
         ballist.append('80001-100000')
    else:
         ballist.append('100001-')
```
